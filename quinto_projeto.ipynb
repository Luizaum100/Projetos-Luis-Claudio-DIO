```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow

# Carrega a imagem
img = cv2.imread('scary-movie.jpg')

# Verifica se a imagem foi carregada corretamente
if img is None:
    print("Erro: Não foi possível carregar a imagem. Verifique o nome e o caminho do arquivo.")
else:
    img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    cv2_imshow(img_rgb)
```

```python
from mtcnn.mtcnn import MTCNN

# Cria o detector de faces
detector = MTCNN()

# Detecta as faces na imagem
faces = detector.detect_faces(img_rgb)

# Desenha um retângulo ao redor de cada face detectada
for face in faces:
    x, y, width, height = face['box']
    cv2.rectangle(img, (x, y), (x+width, y+height), (0, 255, 0), 2)

# Mostra a imagem com as faces detectadas
cv2_imshow(img)
```
